```python
#Combining what I currently have and matches based on the current name
#Note that I have three frames right now, pivot_df, df_revenue, and code_df
merge_1 = pd.merge(pivot_df, code_df, left_on='Organization', right_on='Name')
merge_1.head(2)
```

```python
#Merging in the next info
df_rev_merge = pd.merge(merge_1, df_revenue, left_on='source_org_id', right_on='Company Name')
df_rev_merge = df_rev_merge.drop(columns = ['Organization_ID', 'source_org_id', 'Name'])
df_rev_merge.drop(df_rev_merge.columns[[0]], axis = 1)

df_rev_merge
```
